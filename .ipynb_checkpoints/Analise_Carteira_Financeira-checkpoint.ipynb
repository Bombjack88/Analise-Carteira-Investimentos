# Importar as bibliotecas


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

# Inputs: carteira de ativos e data de hoje 

```python
# Importar a minha Carteira
carteira = pd.read_excel('Carteira_Ativos.xlsx')
display(carteira)
Data_hoje='2023-01-28'
```

# Criar um DataFrame das cotações dos Ativos

```python
cotacoes_carteira = pd.DataFrame()

for ativo in carteira['Ativos']:
    cotacoes_carteira[ativo]= yf.download('{}'.format(ativo), start='2020-01-01',  end='{}'.format(Data_hoje), progress=False)['Adj Close']
    
display(cotacoes_carteira)
```

```python
# Analisar a informação do DataSet
cotacoes_carteira.info()
```

# Comportamento individual das ações

```python
carteira_norm = cotacoes_carteira / cotacoes_carteira.iloc[0]
carteira_norm.plot(figsize=(15, 5))
plt.legend(loc='upper left')
```

# Total da Carteira

```python
valor_investido = pd.DataFrame()

for ativo in carteira['Ativos']:
    valor_investido[ativo] = cotacoes_carteira[ativo] * carteira.loc[carteira['Ativos']==ativo, 'Qtde'].values[0]

#  Valor total da carteira

valor_investido['Total'] = valor_investido.sum(axis=1)

display(valor_investido)
```

```python
#Carteira e Componentes - > 2022-02-17 (data do primeiro investimento na NetFlix)
valor_investido['2022-02-17':].plot(figsize=(15, 5))
plt.legend()
plt.show()
```

```python
# Apenas Carteira
valor_investido['Total']['2022-02-17':].plot(figsize=(15, 5))
plt.legend()
plt.show()
```

```python
Valor_Carteira =valor_investido['Total'][-1]

print(f'O Valor da carteira é de {Valor_Carteira:.2F}.')
```

```python
retorno_carteira = valor_investido['Total'][-1] / valor_investido['Total']['2022-02-17':][0] - 1
print('Retorno da Carteira: {:.2%}'.format(retorno_carteira))
```

```python
#Compara Valores
print(f" Valor inicial 2022/02/17: {valor_investido['Total']['2022-02-17':][0]:.2F}")
print(f" Valor inicial 02/01/2020: {valor_investido['Total'][0]:.2F}")
```
